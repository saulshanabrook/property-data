# Near Greenfield

```python
import pandas as pd

pd.options.display.max_columns = None

from load_data import df, HIDE_USE_DESCS, HIDE_OWNERS
from distance_matrix import add_directions_columns
```

```python
ALLOWED_TOWNS = ["GREENFIELD"]
filtered = add_directions_columns(df.query(
    'CITY == @ALLOWED_TOWNS and USE_DESC != @HIDE_USE_DESCS and OWNER1 != @HIDE_OWNERS and BLDG_VAL < 200000'
).groupby("OWNER1").filter(lambda x: x.LOT_SIZE.sum() > 10))
```

```python
close_to_town = filtered[filtered['distance'] < 3].groupby("OWNER1").filter(lambda x: x.LOT_SIZE.sum() > 10)
close_to_town.explore(
    popup=True,
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='ESRI',
    column='OWNER1'
)
```

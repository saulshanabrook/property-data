```python
import pandas as pd

pd.options.display.max_columns = None

from load_data import df, HIDE_USE_DESCS, HIDE_OWNERS
from distance_matrix import add_directions_columns
```

```python
ALLOWED_TOWNS = ["greenfield"]
filtered = add_directions_columns(df.query(
    'CITY.str.lower() == @ALLOWED_TOWNS'
))
```

```python
STREETS = ["VALLEY VIEW DR", "SUNRISE AVE", "ESTER AVE", "STONE RIDGE LN"]
```

```python
ten_minutes_walk = filtered.query("FULL_STR == @STREETS")
res = ten_minutes_walk.explore(
    popup=True,
    tiles='https://server.arcgisonline.com/ArcGIS/rest/services/World_Imagery/MapServer/tile/{z}/{y}/{x}',
    attr='ESRI',
    column='FULL_STR'
)
```

```python
ten_minutes_walk.sort_values("FULL_STR")[["SITE_ADDR", "OWNER1", "USE_DESC"]].to_csv("tmp.csv")
```
